```python
import matplotlib.pyplot as plt
from scipy import stats

x = data['2006-07']
y = data['2007-08']

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
	return slope * x + intercept


mymodel = list(map(myfunc, x))

plt.scatter(x, y)

plt.plot(x, mymodel)

plt.show()
```
